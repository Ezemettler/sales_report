```python
import pandas as pd
```

```python
data_electronics = pd.read_csv('electronics_sales.csv')
data_electronics.head(15)
```

```python
conteo_productos = data_electronics['Product'].value_counts()
print(conteo_productos)
```

```python
productos = data_electronics.groupby('Product').mean()
productos
```

```python
df_electronics = data_electronics
```

```python
# Reemplazamos productos 'LG Washing Machine', 'LG Dryer' y 'Vareebadd Phone' a productos que se vendan más relacionados a smartphones y laptops.
df_electronics['Product'] = df_electronics['Product'].replace('LG Washing Machine', 'Samsung Galaxy Phone')
df_electronics['Product'] = df_electronics['Product'].replace('LG Dryer', 'Dell Inspiron Laptop')
df_electronics['Product'] = df_electronics['Product'].replace('Vareebadd Phone', 'HP Pavilion Laptop')
df_electronics['Product'] = df_electronics['Product'].replace('Flatscreen TV', 'Motorola Edge Phone')

# Selecciona las filas con nombre de producto y les asigna la marca.
df_electronics.loc[df_electronics['Product'] == 'iPhone', 'brand'] = 'Apple'
df_electronics.loc[df_electronics['Product'] == 'Macbook Pro Laptop', 'brand'] = 'Apple'
df_electronics.loc[df_electronics['Product'] == 'Apple Airpods Headphones', 'brand'] = 'Apple'
df_electronics.loc[df_electronics['Product'] == 'USB-C Charging Cable', 'brand'] = 'Apple'
df_electronics.loc[df_electronics['Product'] == 'Bose SoundSport Headphones', 'brand'] = 'Bose'
df_electronics.loc[df_electronics['Product'] == 'Google Phone', 'brand'] = 'Google'
df_electronics.loc[df_electronics['Product'] == 'Lightning Charging Cable', 'brand'] = 'Generic'
df_electronics.loc[df_electronics['Product'] == 'AAA Batteries (4-pack)', 'brand'] = 'Duracell'
df_electronics.loc[df_electronics['Product'] == 'AA Batteries (4-pack)', 'brand'] = 'Duracell'
df_electronics.loc[df_electronics['Product'] == 'Wired Headphones', 'brand'] = 'Sony'
df_electronics.loc[df_electronics['Product'] == '27in FHD Monitor', 'brand'] = 'Samsung'
df_electronics.loc[df_electronics['Product'] == '27in 4K Gaming Monitor', 'brand'] = 'Asus'
df_electronics.loc[df_electronics['Product'] == '34in Ultrawide Monitor', 'brand'] = 'ViewSonic'
df_electronics.loc[df_electronics['Product'] == 'ThinkPad Laptop', 'brand'] = 'Lenovo'
df_electronics.loc[df_electronics['Product'] == '20in Monitor', 'brand'] = 'Dell'
df_electronics.loc[df_electronics['Product'] == 'Samsung Galaxy Phone', 'brand'] = 'Samsung'      
df_electronics.loc[df_electronics['Product'] == 'Dell Inspiron Laptop', 'brand'] = 'Dell'                   
df_electronics.loc[df_electronics['Product'] == 'HP Laptop', 'brand'] = 'HP'
df_electronics.loc[df_electronics['Product'] == 'Motorola Edge Phone', 'brand'] = 'Motorola'

# Imprime el dataframe restante luego de modificaciones
df_electronics.head(10)
```
